## Multi-Head Attention From Scratch

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# creating single attention head

class AttentionHead(nn.Module):

    def __init__(self,
                 embed_dim,
                 head_dim):

        super().__init__()

        self.query = nn.Linear(
            embed_dim,
            head_dim,
            bias=False
        )

        self.key = nn.Linear(
            embed_dim,
            head_dim,
            bias=False
        )

        self.value = nn.Linear(
            embed_dim,
            head_dim,
            bias=False
        )

    def forward(self, x):

        Q = self.query(x)

        K = self.key(x)

        V = self.value(x)

        scores = Q @ K.transpose(-2,-1)

        scores = scores / (K.shape[-1] ** 0.5)

        weights = F.softmax(
            scores,
            dim=-1
        )

        out = weights @ V

        return out
```

```python
embed_dim = 256
num_heads = 8

head_dim = embed_dim // num_heads
```

```python
# MultiHead
class MultiHeadAttention(nn.Module):

    def __init__(self,
                 embed_dim,
                 num_heads):

        super().__init__()

        head_dim = embed_dim // num_heads

        self.heads = nn.ModuleList(

            [
                AttentionHead(
                    embed_dim,
                    head_dim
                )

                for _ in range(num_heads)
            ]
        )

        self.proj = nn.Linear(
            embed_dim,
            embed_dim
        )

    def forward(self, x):

        outputs = [

            head(x)

            for head in self.heads

        ]

        out = torch.cat(
            outputs,
            dim=-1
        )

        out = self.proj(out)

        return out
```

## Self-Attention

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
batch_size = 1
seq_len = 4
embed_dim = 8

# random embedding
X = torch.randn(
    batch_size,
    seq_len,
    embed_dim
)

print(X.shape)
```

### Transformers learns following 3 metrics

```python
query = nn.Linear(
    embed_dim,
    embed_dim,
    bias=False
)

key = nn.Linear(
    embed_dim,
    embed_dim,
    bias=False
)

value = nn.Linear(
    embed_dim,
    embed_dim,
    bias=False
)
```

```python
# generating metrics for the X
Q = query(X)
K = key(X)
V = value(X)
```

```python
print(Q.shape)
print(K.shape)
print(V.shape)
```

### calculate attension score = Q × K^T

```python
scores = Q @ K.transpose(-2,-1)

print(scores.shape)
```

```python
print(scores[0])
```

```python
# scale score
d_k = K.shape[-1]

scores = scores / (d_k ** 0.5)
```

```python
# convert score into probability
weights = F.softmax(
    scores,
    dim=-1
)

print(weights[0])
```

```python
# now create context vector
output = weights @ V
print(output.shape)
```

# Full implementation of Self attention

```python
class SelfAttention(nn.Module):

    def __init__(self, embed_dim):

        super().__init__()

        self.query = nn.Linear(
            embed_dim,
            embed_dim,
            bias=False
        )

        self.key = nn.Linear(
            embed_dim,
            embed_dim,
            bias=False
        )

        self.value = nn.Linear(
            embed_dim,
            embed_dim,
            bias=False
        )

    def forward(self, x):

        Q = self.query(x)

        K = self.key(x)

        V = self.value(x)

        scores = (
            Q @ K.transpose(-2,-1)
        )

        scores = (
            scores /
            (K.shape[-1] ** 0.5)
        )

        weights = F.softmax(
            scores,
            dim=-1
        )

        out = weights @ V

        return out
```
